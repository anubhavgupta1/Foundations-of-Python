#Lambda Operator

```
 f = lambda x, y : x + y
 f(1,1)
```

```
f(2,4)
```

```
foo = [2, 18, 9, 22, 17, 24, 8, 12, 27]
```

```
# filter() function

divisible_by_3 = filter(lambda x: x % 3 == 0, foo)
print(list(divisible_by_3))
```

```
# map() function

lambda_map = map(lambda x: x * 2 + 10, foo)
print(list(lambda_map))
```

```
# reduce() function

import functools
lambda_reduce = functools.reduce(lambda x, y: x + y, foo)
print(lambda_reduce)
```
